```python
import pandas as pd
import os
import glob

# Set the path to the directory containing the CSV files
path = './data/summary'

# Use glob to list all CSV files in the directory
csv_files = glob.glob(os.path.join(path, '*.csv'))

# Dictionary to hold dataframes with keys as filenames
dataframes = {}

# Read each CSV file and store it in the dictionary with the filename as the key
for file in csv_files:
    # Extract the filename without the path and the extension
    filename = os.path.splitext(os.path.basename(file))[0]
    dataframes[filename] = pd.read_csv(file)

# Access each dataframe using the filename as the key, e.g., dataframes['file1']
print("DataFrames loaded:", list(dataframes.keys()))
```
